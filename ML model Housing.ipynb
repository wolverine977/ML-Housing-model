```python
label=df.median_house_value.copy()
```

```python
feature=df.drop(columns=["median_house_value"])
```

```python
feature.info()
```

```python
feature.select_dtypes("float")
```

```python
import scipy.stats as stats
```

```python
feat1=feature.select_dtypes("float").apply(lambda x: stats.zscore(x))
```

```python
feat1   #zcsore measures how many standard deviations a data point is from the mean in a distribution.
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
feat1.agg(["mean","std"])     #thats exactly we done above standarizing feature so we get mean 0 and std 1
```

```python
feature.ocean_proximity.value_counts()
```

```python
dummies=pd.get_dummies(feature.ocean_proximity)
```

```python
dummies
```

```python
feature=pd.concat([feat1,dummies,df.income_cat], axis=1)
```

```python
feature
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
train_set,test_set=train_test_split(feature, test_size=0.2, random_state=123)
```

```python
feature.income_cat.value_counts(normalize=True)
```

```python
train_set.income_cat.value_counts(normalize=True)
```

```python
test_set.income_cat.value_counts(normalize=True)
```

```python
train_set=train_set.sample(frac=1, random_state=123)
train_set
```

```python
train_set.drop(columns=["income_cat"],inplace=True)
test_set.drop(columns=["income_cat"],inplace=True)
```

```python
train_y=label.loc[train_set.index]
test_y=label.loc[test_set.index]
```

```python
train_y
```

## Training our model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_for= RandomForestRegressor(random_state=42, max_features="sqrt", max_depth=75, min_samples_split=2, n_estimators=500)
```

```python
model_for.fit(train_set, train_y)
```

```python
model_for.score(train_set, train_y)       #return the coefficient of determination of the prediction it seems like our model is overfitting
```

```python
pred=model_for.predict(train_set)    #predictions
pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse= mean_squared_error(train_y, pred)
rmse= np.sqrt(mse)
rmse
```
